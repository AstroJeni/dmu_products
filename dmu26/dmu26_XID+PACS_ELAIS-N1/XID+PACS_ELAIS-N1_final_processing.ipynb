# Final Processing
The final processing stage requires:
1. Quick validation of catalogues and Bayesian Pvalue maps
2. Skewness level
3. Adding flag to catalogue

```python
import seaborn as sns
from astropy.table import Table
%matplotlib inline
import numpy as np
import pylab as plt
```

## Read tables

```python
SERVS_cat=Table.read('../data/ELAIS_N1/PACS/dmu26_XID+PACS_ELAIS-N1_SERVS_cat_20170801.fits')
```

## Look at Symmetry of PDFs to determine depth level of catalogue

```python
skew=(SERVS_cat['FErr_PACS_100_u']-SERVS_cat['F_PACS_100'])/(SERVS_cat['F_PACS_100']-SERVS_cat['FErr_PACS_100_l'])
skew.name='(84th-50th)/(50th-16th) percentile'
g=sns.jointplot(x=np.log10(SERVS_cat['F_PACS_100']),y=skew, kind='scatter')
```

For 100 $\mathrm{\mu m}$ depth is ~ 4mJy

```python
skew=(SERVS_cat['FErr_PACS_160_u']-SERVS_cat['F_PACS_160'])/(SERVS_cat['F_PACS_160']-SERVS_cat['FErr_PACS_160_l'])
skew.name='(84th-50th)/(50th-16th) percentile'
g=sns.jointplot(x=np.log10(SERVS_cat['F_PACS_160']),y=skew, kind='scatter')
```

```python
10.0**1.1
```

For 160 $\mathrm{\mu m}$ depth is ~ 4mJy

## Add flag to catalogue

```python
from astropy.table import Column
SERVS_cat.add_column(Column(np.zeros(len(SERVS_cat), dtype=bool),name='flag_PACS_100'))
SERVS_cat.add_column(Column(np.zeros(len(SERVS_cat), dtype=bool),name='flag_PACS_160'))

```

```python
ind_SERVS_100=(SERVS_cat['Pval_res_100']>0.5) | (SERVS_cat['F_PACS_100'] < 12.5)
ind_SERVS_160=(SERVS_cat['Pval_res_160']>0.5) | (SERVS_cat['F_PACS_160'] < 17.5)
```

```python
print(ind_SERVS_100.sum(),ind_SERVS_160.sum(),len(SERVS_cat))
```

```python
SERVS_cat['flag_PACS_100'][ind_SERVS_100]=True
SERVS_cat['flag_PACS_160'][ind_SERVS_160]=True
```

```python
SERVS_cat.write('../data/ELAIS_N1/PACS/dmu26_XID+PACS_ELAIS-N1_SERVS_cat_20170801.fits', format='fits',overwrite=True)
```
